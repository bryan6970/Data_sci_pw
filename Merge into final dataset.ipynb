```python
import pyforest
import pandas as pd
import numpy as np
```

```python
holiday_df = pd.read_excel("Datasets/Holiday.xlsx")
demand_df = pd.read_csv("Datasets/Electricity price and demand/Cleaned singapore electricity usage and demand.csv")
population_df = pd.read_csv("Datasets/Population/stats.csv")
days_df = pd.read_excel("Datasets/Days.xlsx")
```

# Clean holiday df

```python
day_map = {
    "Monday": 1,
    "Tuesday":2,
    "Wednesday": 3,
    "Thursday":4,
    "Friday":5,
    "Saturday":6,
    "Sunday":7
}
```

```python
holiday_df
```

```python
holiday_df = holiday_df.T.reset_index().T
```

```python
# Set empty column names
holiday_df.columns = ['DateTime', 'Day', "Holiday"]
```

```python
holiday_df["DateTime"] = pd.to_datetime(holiday_df["DateTime"], format='%B %d, %Y')
```

```python
holiday_df.set_index("DateTime", inplace=True)
```

```python
holiday_df["Day"] = holiday_df["Day"].map(day_map)
```

```python
population_df.set_index("DateTime", inplace=True)
```

```python
# Drop all columns except the selected column
selected_column = 'Total Population (Number)'
columns_to_drop = [col for col in population_df.columns if col != selected_column]
population_df.drop(columns=columns_to_drop, inplace=True)
```
